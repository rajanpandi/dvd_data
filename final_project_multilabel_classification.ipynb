```python
x=df[["customer_id","rating"]].values
y = df[["name", "store_id"]].values
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical



# Initialize the label encoders for each column
label_encoders = [LabelEncoder() for _ in range(y.shape[1])]

# Apply Label Encoding to each column
for i in range(y.shape[1]):
    y[:, i] = label_encoders[i].fit_transform(y[:, i])

# Now, y contains integer encoded values, so we can one-hot encode it
# If needed, apply one-hot encoding
y = [to_categorical(y[:, i]) for i in range(y.shape[1])]


```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Concatenate the one-hot encoded columns to make a single y array
y = np.concatenate(y, axis=1)

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x_train.shape,y_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf

# Create the model
model = Sequential()

# Input layer with reduced neurons and ReLU activation
model.add(Dense(128, input_dim=2, activation='relu'))  # Adjust input_dim as per x_train

# Hidden layers with ReLU activation
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))

# Output layer with softmax for multi-class classification
model.add(Dense(18, activation="softmax"))

# Compile the model with categorical crossentropy and additional metrics
model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=[tf.keras.metrics.F1Score, "accuracy",
                       tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

# Model summary to see the architecture
model.summary()
```

```python
result = model.fit(x_train, y_train, epochs=5,validation_split=.2,batch_size=200 )
```
